```python
# %%
import urllib.request
from urllib.error import URLError, HTTPError, ContentTooShortError
import urllib.parse
import re
import requests
from bs4 import BeautifulSoup 
import pandas as pd
from datetime import datetime
```

```python
# %%


def download(url):
    print('Downloading:', url)
    try:
        response = urllib.request.urlopen(url)
        header = response.headers
        charset = header.get_content_charset(failobj='utf-8')
        html = response.read().decode(charset)
    except (URLError, HTTPError, ContentTooShortError) as e:
        print('Download error:', e.reason)
        html = None
    return header, html
```

```python
dicts = []
for html in htmls:
    dic_df = {}
    soup = BeautifulSoup(html, 'html5lib')
    columns = soup.find_all('label', class_='readonly')
    image_link = soup.find('td', class_='w2p_fw').img
    values = soup.find_all('td', class_='w2p_fw')
    for i in range(len(columns)):
        if i == 0:
            column = columns[0].text
            value = image_link['src']
            dic_df[column] = value 
        else:
            column = columns[i].text
            value = values[i].text
            dic_df[column] = value
    dic_df['timestamp'] = datetime.now()
    dicts.append(dic_df)
    
df = pd.DataFrame(dicts)
display(df.head())
df.to_csv('coutries.csv')
```
